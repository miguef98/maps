```python
import json
from shapely.geometry import shape, mapping
from shapely.affinity import scale
```

```python

# Load GeoJSON file
with open("provincias_sin_antartida-2.geojson", "r") as f:
    geojson_data = json.load(f)

# Extract features
features = geojson_data["features"]

# Define the scaling factor
scale_factor = 2.0  # Example: Scale the polygon by a factor of 2

features
```

```python

# Scale each feature
for feature in features:
    if feature['properties']['name'] == 'Ciudad Autónoma de Buenos Aires':
        geom = shape(feature["geometry"])
        
        # Compute centroid
        centroid = geom.centroid
        
        # Apply scaling
        scaled_geom = scale(geom, xfact=scale_factor, yfact=scale_factor, origin=(centroid.x, centroid.y))
        
        # Update the geometry in the feature
        feature["geometry"] = mapping(scaled_geom)

# Save the updated GeoJSON
with open("output.geojson", "w") as f:
    json.dump(geojson_data, f, indent=2)
```
